* Why can a CFG production only have a single non-terminal on the LHS? Answer both why it must be singular, and why it must be a non-terminal.

Singular: To be "Context-Free," the rule must apply regardless of context. Multiple symbols on the LHS would imply a context-sensitive rule (e.g., a rule that only applies when an NP is followed by a VB).

Non-Terminal: The LHS represents a high-level structure (like a phrase) that is being expanded. Terminals are final outputs (words) and cannot be expanded further.

* Why is recursion essential in CFGs for modeling natural language? Give a simple example involving a noun phrase or verb phrase.

* Clitics are a special type of syntactic headache. Unlike affixes, which attach at the word level, clitics can attach at the phrase level. For example: The man who saw the bird's camera was not quick enough.'' or Those of us who lived through the '90s've experienced a world without the internet.'' Explain why phrase-level clitic attachment is problematic for a CFG, and discuss how (if at all) a CFG could be adapted to model this behavior.

Problem: CFGs generally assume that constituents are contiguous and that markers (like possessive 's) attach to the head word of the phrase they modify. Phrase-level attachment (like attaching 's to "The man who saw the bird") breaks the standard assumption that the suffix belongs to the immediate word ("bird") rather than the whole phrase ("The man...").

Adaptation: A CFG could potentially model this by treating the clitic as a separate terminal item that attaches to the end of a specific phrase type, e.g., NP_Possessive -> NP 's.
